```python
import cirq
from cirq.contrib.svg import SVGCircuit
import matplotlib.pyplot as plt

qc = cirq.Circuit()
q0,q1,q2,q3 = cirq.LineQubit.range(4)

qc.append(cirq.X(q0))
qc.append(cirq.X(q2))

# -------------------- Direct QFT ----------------------
#qc.append(cirq.qft(q0,q1,q2,q3,without_reverse=True))

qc.append(cirq.H(q0))
qc.append(cirq.CZ(q0,q1)**(1/2))
qc.append(cirq.CZ(q0,q2)**(1/4))
qc.append(cirq.CZ(q0,q3)**(1/8))

qc.append(cirq.H(q1))
qc.append(cirq.CZ(q1,q2)**(1/2))
qc.append(cirq.CZ(q1,q3)**(1/4))

qc.append(cirq.H(q2))
qc.append(cirq.CZ(q2,q3)**(1/2))

qc.append(cirq.H(q3))
# -----------------------------------------------------

# --------------------- Inverse QFT -------------------
#qc.append(cirq.qft(q0,q1,q2,q3,inverse=True,without_reverse=True))

qc.append(cirq.H(q3))

qc.append(cirq.CZ(q2,q3)**(-1/2))
qc.append(cirq.H(q2))

qc.append(cirq.CZ(q1,q3)**(-1/4))
qc.append(cirq.CZ(q1,q2)**(-1/2))
qc.append(cirq.H(q1))

qc.append(cirq.CZ(q0,q3)**(-1/8))
qc.append(cirq.CZ(q0,q2)**(-1/4))
qc.append(cirq.CZ(q0,q1)**(-1/2))
qc.append(cirq.H(q0))
# -----------------------------------------------------

qc.append(cirq.measure(q3,q2,q1,q0))

print()
print(qc)
print()

s = cirq.Simulator()
samples = s.run(qc, repetitions=1000)

cirq.plot_state_histogram(samples, plt.subplot())
plt.show()

SVGCircuit(qc)
```
